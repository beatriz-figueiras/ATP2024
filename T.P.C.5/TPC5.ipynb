```python
print("Bem-vindo!")

def listar(cinema):
    print("-----Filmes em exibição-----")
    for sala in cinema:
        print(f"Filme: {sala[2]} | capacidade: {sala[0]} lugares")

def disponivel(cinema, filme, lugar):
    for sala in cinema:
        if sala[2] == filme:
            if lugar in sala[1]:
                return False
            return True
    return False

def vendebilhete(cinema, filme, lugar):
    for i, sala in enumerate(cinema):
        if sala[2] == filme:
            if lugar in sala[1]:
                print("Lugar ocupado.")
                return cinema
            if lugar > sala[0]:
                print("Lugar inválido.")
                return cinema
            cinema[i] = (sala[0], sala[1] + [lugar], sala[2])
            print(f"Bilhete vendido para o filme {filme}, lugar {lugar}")
    print("Filme não encontrado.")
    return cinema

def listarDisp(cinema):
    print("-----Disponibilidade-----")
    for sala in cinema:
        disponivel = sala[0] - len(sala[1])
        print(f"Filme: {sala[2]} | lugares disponíveis: {disponivel} / {sala[0]}")

def inserSala(cinema, sala):
    for hSala in cinema:
        if hSala[2] == sala[2]:
            print(f"Já existe a sala para {sala[2]}.")
            return cinema
    cinema.append(sala)
    print(f"Sala com filme {sala[2]} adicionada com sucesso.")
    return cinema

def main():
    cinema1 = []
    continuar = True

    while continuar:
        escolha = int(input(
            "O que gostaria de fazer?\n"
            "[1] Listar filmes\n"
            "[2] Verificar disponibilidade\n"
            "[3] Vender bilhete\n"
            "[4] Listar disponibilidade\n"
            "[5] Inserir nova sala\n"
            "[0] Sair\n"    
        ))

        if escolha == 0:
            print("Saiu.")
            continuar = False
        
        elif escolha == 1:
            listar(cinema1)
        
        elif escolha == 2:
            filme = input("Qual filme pretende verificar?")
            lugar = int(input("Qual lugar pretende verificar?"))
            if disponivel(cinema1, filme, lugar):
                print("Lugar disponível.")
            else:
                print("Lugar não disponível.")
        
        elif escolha == 4:
            listarDisp(cinema1)
        
        elif escolha == 5:
            capacidade = int(input("Qual a capacidade da sala?"))
            filme = input("Que filme será exibido nesta sala?")
            sala = (capacidade, [], filme)
            cinema1 = inserSala(cinema1, sala)
        
        else:
            print("Opção inválida.")

main()
```
