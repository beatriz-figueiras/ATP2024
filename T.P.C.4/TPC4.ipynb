```python
import random

def monitor():
    print("\nMenu:")
    print("(1) Criar Lista Aleatória")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Está Ordenada por Ordem Crescente")
    print("(8) Está Ordenada por Ordem Decrescente")
    print("(9) Procurar um Elemento")
    print("(0) Sair")

def criarLista():
    lista = []
    N = int(input("Digite quantos números pretende ter na sua lista: "))
    for i in range(N):
        num = random.randint(1,100)
        lista.append(num)
    return lista

def lerLista():
    lista = []
    N = int(input("Digite quantos números pretende ter na sua lista: "))
    for i in range(N):
        num = int(input("Introduza o número " + str(i) + ":"))
        lista.append(num)
        i = i + 1
    return lista

def soma(lista):
    soma = 0
    for num in lista:
        soma = soma + num
    return soma

def media(lista):
    soma = 0
    for num in lista:
        soma = soma + num
        media = soma / len(lista)
    return media

def maiorElem(lista):
    maior = lista [0]
    for num in lista:
        if num > maior:
            maior = num
    return maior

def menorElem(lista):
    menor = lista [0]
    for num in lista:
        if num < menor:
            menor = num
    return menor

def estaOrdeCres(lista): 
    listaOrdenada = lista.copy()
    listaOrdenada.sort()
    if lista == listaOrdenada:
        print("Sim. A lista está ordenada por ordem crescente.")
    else:
        print("Não. A lista não está ordenada por ordem crescente.")

def estaOrdeDecres(lista): 
    listaOrdenada = lista.copy()
    listaOrdenada.sort()
    listaOrdenada.reverse()
    if lista == listaOrdenada:
        print("Sim. A lista está ordenada por ordem decrescente.")
    else:
        print("Não. A lista não está ordenada por ordem decrescente.")

def procurarElem(lista):
    elem = int(input("Digite o elemento que procura:"))
    posicao = lista.index(elem) if elem in lista else -1
    if posicao != -1:
        print(f"O elemento {elem} foi encontrado na posição: {posicao}.")
    else:
        print("O elemento não foi encontrado na lista.")

def menu():
    lista = []
    continuar = True
    monitor()
    while continuar:
        
        opcao = input("Escolha uma opção:")
        print(f"Opção({opcao})")
        if opcao in {'3','4', '5', '6', '7', '8', '9'} and not lista:
            print("A lista está vazia. Por favor, comece por escolher a opção 1 ou 2.")
            menu()
            return
        if opcao == '1':
            lista = criarLista()
            print("Lista aleatória:", lista)
        elif opcao == '2':
            lista = lerLista()
            print("Lista: ", lista)
        elif opcao == '3':
            print(soma(lista))
        elif opcao == '4': 
            print(media(lista))
        elif opcao == '5':
            print(maiorElem(lista))
        elif opcao == '6':
            print(menorElem(lista))
        elif opcao == '7':
            estaOrdeCres(lista)
        elif opcao == '8':
            estaOrdeDecres(lista)
        elif opcao == '9':
            procurarElem(lista)
        elif opcao == '0':
            print("Sair")
            continuar = False
        else:
            print("Opção inválida. Por favor, escolha uma opção válida")

menu()
```
